```python
###Duration of dataset
```

```python
import wave
import os

path = "/home/phannet.pov/tacotron2_khmer/khmer-dataset-female/wavs"
duration_seconds = 0
for x in os.listdir(path):
    if x.endswith(".wav"):
        file = path+"/"+x
        with wave.open(file) as wav:
            duration_seconds = (wav.getnframes() / wav.getframerate())+duration_seconds
            # print(x)

seconds = duration_seconds
seconds = seconds % (24 * 3600)
hour = seconds // 3600
seconds %= 3600
minutes = seconds // 60
seconds %= 60
print("khmer-female-dataset")
print("Total duration of dataset: %d:%02d:%02d" % (hour, minutes, seconds))
```

```python
#READ TEXT FIEL

path='/home/phannet.pov/tacotron2_khmer/khmer-dataset/text/list.txt'

file1 = open(path, 'r')
Lines = file1.readlines()
 
count = 0
# Strips the newline character
for line in Lines:
    count += 1
    #read only text in each line
    print("Line{}: {}".format(count, line.strip()))
    #count number of word in each llines
    print("Count number of words in Line{}: {}".format(count, len(line.split())))
```

```python
#summation all the words 
sum = 0
for line in Lines:
    sum = sum + len(line.split())
print("Summation all the words:  ",sum)
```

```python
from pathlib import Path
from scipy.io import wavfile
path_of_the_directory = '/home/phannet.pov/tacotron2_khmer/khmer-dataset/wavs'
print("Files and directories in a specified path:")
file = Path(path_of_the_directory ).glob('*')
for i in file:
  fs, data = wavfile.read(i)
  print("\nFile_name: ",i)
  print("Sampling Rate: ",fs)
  print(data)
```
